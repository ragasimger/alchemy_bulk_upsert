```python
import numpy as np
import pandas as pd
```

```python
df = pd.read_csv('datas/mock_data.csv')
fields = df.columns.to_list()
first_name_list = df['first_name'].to_list()
last_name_list = df['last_name'].to_list()
email_list = df['email'].to_list()
gender_list = df['gender'].to_list()
ip_address_list = df['ip_address'].to_list()
phone_list = df['phone'].to_list()
father_name_list = df['father_name'].to_list()
grandfather_name_list = df['grandfather_name'].to_list()
mother_name_list = df['mother_name'].to_list()
```

```python
custom_dtype = np.dtype([
    ('non_primary_id', np.int32),
    ('first_name', 'U50'),
    ('last_name', 'U50'),
    ('email', 'U50'),
    ('gender', 'U50'),
    ('ip_address', 'U50'),
    ('phone', 'U50'),
    ('father_name', 'U50'),
    ('grandfather_name', 'U50'),
    ('mother_name', 'U50'),
])
```

```python
num_samples = 100000
non_primary_id = [i for i in range(1, num_samples+1)]
```

```python
dict_test_data = {
    'non_primary_id': np.array(non_primary_id, dtype='U50'),
    'first_name': np.random.choice(first_name_list, size=num_samples),
    'last_name': np.random.choice(last_name_list, size=num_samples),
    'email': np.random.choice(email_list, size=num_samples),
    'gender': np.random.choice(gender_list, size=num_samples),
    'ip_address': np.random.choice(ip_address_list, size=num_samples),
    'phone': np.random.choice(phone_list, size=num_samples),
    'father_name': np.random.choice(father_name_list, size=num_samples),
    'grandfather_name': np.random.choice(grandfather_name_list, size=num_samples),
    'mother_name': np.random.choice(mother_name_list, size=num_samples),
}
```

```python
structured_array = np.empty(num_samples, dtype=custom_dtype)
for field_name in custom_dtype.names:
    # print("Processing field:", field_name)
    structured_array[field_name] = dict_test_data[field_name]
# structured_array
```

```python
# structured_dict = {field_name: dict_test_data[field_name].astype(custom_dtype[field_name]) for field_name in custom_dtype.names}
structured_dict = {
    field_name: np.array(dict_test_data[field_name], dtype=custom_dtype[field_name])
    for field_name in custom_dtype.names
}
```

```python
df = pd.DataFrame(structured_dict)
# df.to_csv('insert_df_100k.csv', index=False)
df
```
